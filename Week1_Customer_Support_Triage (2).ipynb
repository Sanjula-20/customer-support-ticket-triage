```python
def clean_text(text):
    text = str(text).lower()          # convert to lowercase
    text = re.sub(r'[^a-z\s]', ' ', text)  # remove special characters
    text = re.sub(r'\s+', ' ', text)  # remove extra spaces
    return text.strip()
```

```python
df["cleaned_text"] = df["Ticket Description"].apply(clean_text)
```

```python
df[["Ticket Description", "cleaned_text"]].head(5)
```

```python
def classify_issue(text):
    text = str(text).lower()

    if any(word in text for word in ["refund", "money back", "return"]):
        return "REFUND"

    elif any(word in text for word in ["payment", "billing", "charged", "transaction"]):
        return "PAYMENT"

    elif any(word in text for word in ["login", "password", "otp", "signin", "sign in"]):
        return "LOGIN"

    elif any(word in text for word in ["delivery", "shipping", "shipment", "courier"]):
        return "DELIVERY"

    elif any(word in text for word in ["error", "bug", "crash", "not working", "issue"]):
        return "BUG"

    else:
        return "GENERAL"
```

```python
df["issue_category"] = df["cleaned_text"].apply(classify_issue)
```

```python
df[["Ticket Description", "issue_category"]].head(10)
```

```python
def assign_priority(text):
    text = str(text).lower()

    # P0 - Critical (service completely blocked)
    if any(word in text for word in [
        "urgent", "asap", "failed", "not working",
        "system down", "service down",
        "payment failed", "transaction failed",
        "money deducted", "cannot place order"
    ]):
        return "P0"

    # P1 - High (major error but not total failure)
    elif any(word in text for word in [
        "error", "issue", "problem", "unable",
        "bug", "crash", "not responding"
    ]):
        return "P1"

    # P2 - Medium (delay / performance issues)
    elif any(word in text for word in [
        "delay", "delayed", "slow", "late",
        "pending", "taking time"
    ]):
        return "P2"

    # P3 - Low (general queries)
    else:
        return "P3"
```

```python
df["priority_level"] = df["cleaned_text"].apply(assign_priority)
```

```python
df[["Ticket Description", "priority_level"]].head(10)
```

```python
from datetime import datetime, timedelta

sla_map = {
    "P0": 2,    # Critical
    "P1": 6,    # High
    "P2": 24,   # Medium
    "P3": 48    # Low
}
```

```python
df["sla_hours"] = df["priority_level"].map(sla_map)
```

```python
df["due_time"] = df["sla_hours"].apply(
    lambda x: datetime.now() + timedelta(hours=x)
)
```

```python
df[["priority_level", "sla_hours", "due_time"]].head(10)
```

```python
print("----- SUPPORT MANAGER REPORT -----\n")

print("Total Tickets:")
print(len(df))

print("\nTickets by Priority Level:")
print(df["priority_level"].value_counts())

print("\nTickets by Issue Category:")
print(df["issue_category"].value_counts())
```

```python
final_df = df[
    [
        "Ticket ID",
        "Ticket Description",
        "issue_category",
        "priority_level",
        "sla_hours",
        "due_time"
    ]
]
```
